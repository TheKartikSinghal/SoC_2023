```python
def ReLU(x):
    if (x>0):
        return x
    else:
        return 0
```

```python
def softmax(z):
    exp_z = np.exp(z)
    sum = exp_z.sum()
    softmax_z = np.round(exp_z/sum,3)
    return softmax_z
```
